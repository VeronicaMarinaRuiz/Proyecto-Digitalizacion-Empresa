```python
# Volvemos a crear la conexi贸n y creamos la tabla xml
cnx = mysql.connector.connect(user = 'root', password = 'AlumnaAdalab',
                              host = '127.0.0.1', database = 'da_proyecto_modulo1')


mycursor = cnx.cursor()
try:
    mycursor.execute("""CREATE TABLE `tabla_xml` (
`index_xml` INT NOT NULL,
`time` VARCHAR(150),
`age` VARCHAR(150),
`gender` VARCHAR(150),
`index_sql` INT NOT NULL,
PRIMARY KEY(`index_xml`),
	CONSTRAINT `fk_sql_xml` 
		FOREIGN KEY(`index_sql`) 
	REFERENCES `tabla_sql`(`index_sql`)
    ON DELETE CASCADE ON UPDATE CASCADE
)""")
    cnx.commit()
    print(mycursor)
except mysql.connector.Error as err:
    print(err)
    print("Error Code:", err.errno)
    print("SQLSTATE", err.sqlstate)
    print("Message", err.msg)
```
